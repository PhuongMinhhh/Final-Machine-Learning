```python
def dummies(x,df):
    temp = pd.get_dummies(df[x], drop_first = True).astype(int)
    df = pd.concat([df, temp], axis = 1)
    df.drop([x], axis = 1, inplace = True)
    return df
# Applying the function to the df

df = dummies('mainroad',df)
df = dummies('guestroom',df)
df = dummies('solarpower',df)
df = dummies('subsol',df)
df = dummies('airconditioning',df)
df = dummies('favorite',df)
df = dummies('status',df)
```

```python
def build_model(X,y):
    X = sm.add_constant(X) #Adding the constant for intercept
    lm = sm.OLS(y,X).fit() # fitting the model
    print(lm.summary()) # model summary
    return X
    
```

```python
def checkVIF(X):
    vif = pd.DataFrame()
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return(vif)
```

```python
# 5)Sau khi huấn luyện xong mô hình thì muốn cải thiện độ chính xác, ta sẽ làm gì để giải quyết nó? Phân tích các trường hợp sai, đề ra giải pháp và thực hiện nó, sau đó đánh giá xem có cải tiến so với trước không. 

# Độ chính xác của một mô hình Machine Learning phụ thuộc vào nhiều yếu tố, bao gồm:

# - Kiến trúc mô hình: Kiến trúc mô hình quyết định cách thức mô hình xử lý dữ liệu và học hỏi. Một kiến trúc mô hình tốt sẽ giúp mô hình phát hiện các mối tương quan phức tạp trong dữ liệu và đưa ra dự đoán chính xác hơn.
# - Dữ liệu huấn luyện: Dữ liệu huấn luyện là nguồn cung cấp thông tin cho mô hình học tập. Dữ liệu huấn luyện càng phong phú và chất lượng thì mô hình càng có khả năng học hỏi tốt hơn và đưa ra dự đoán chính xác hơn.
# - Các hyperparameter của mô hình: Các hyperparameter là các tham số của mô hình có thể điều chỉnh để cải thiện hiệu suất của mô hình. Việc điều chỉnh các hyperparameter phù hợp sẽ giúp mô hình đạt được độ chính xác cao hơn.

# Để cải thiện độ chính xác của mô hình sau khi huấn luyện,  cần phân tích các trường hợp sai, đề ra giải pháp và thực hiện nó, sau đó đánh giá xem có cải tiến so với trước không.

# 1. Phân tích các trường hợp sai

# Bước đầu tiên là phân tích các trường hợp sai của mô hình. Điều này có thể được thực hiện bằng cách xem xét các mẫu dữ liệu mà mô hình đã dự đoán sai. Cần xác định xem mô hình đã sai do các lỗi sau:

# - Lỗi do dữ liệu: Dữ liệu huấn luyện có thể không đủ, không chính xác hoặc không phù hợp với bài toán.
# - Lỗi do kiến trúc mô hình: Kiến trúc mô hình có thể không phù hợp với bài toán hoặc không đủ phức tạp để học được các mối quan hệ phức tạp trong dữ liệu.
# - Lỗi do các hyperparameter: Các hyperparameter của mô hình có thể không được tối ưu hóa.

# 2. Đề ra giải pháp

# Sau khi xác định nguyên nhân gây ra các lỗi, cần đề ra giải pháp để khắc phục. Các giải pháp có thể sử dụng như:

# - Thu thập thêm dữ liệu: Nếu dữ liệu huấn luyện còn hạn chế, có thể thử tăng cường dữ liệu huấn luyện bằng các kỹ thuật như:
# + Data augmentation: Tạo ra các dữ liệu mới từ dữ liệu cũ bằng cách thay đổi các đặc trưng của dữ liệu.
# +Data sampling: Chọn lọc các dữ liệu có chất lượng cao để huấn luyện mô hình.
# - Thay đổi kiến trúc mô hình: Nếu kiến trúc mô hình không phù hợp, cần thay đổi kiến trúc mô hình để phù hợp hơn với bài toán. Ví dụ, nếu bài toán yêu cầu học các mối quan hệ phức tạp, chúng ta có thể sử dụng kiến trúc mô hình sâu hơn.
# - Thay đổi các hyperparameter: Nếu các hyperparameter không được tối ưu hóa, chúng ta có thể thay đổi các hyperparameter để cải thiện hiệu suất của mô hình.
# - Sử dụng các thuật toán ensemble: Sử dụng nhiều mô hình khác nhau để đưa ra dự đoán và sau đó sử dụng một thuật toán để kết hợp các dự đoán này lại với nhau.
# - Sử dụng các kỹ thuật kiểm soát overfitting: Overfitting là hiện tượng mô hình học quá nhiều chi tiết của dữ liệu huấn luyện và dẫn đến việc mô hình không thể tổng quát hóa tốt cho dữ liệu mới. Có nhiều kỹ thuật kiểm soát overfitting khác nhau, chẳng hạn như:
# + Loại bỏ dữ liệu dư thừa: Loại bỏ các dữ liệu không cần thiết trong tập dữ liệu huấn luyện.
# + Giảm kích thước của mô hình: Giảm số lượng nút hoặc tầng trong mô hình.
# + Sử dụng regularization: Sử dụng các thuật toán regularization để hạn chế mô hình học quá nhiều chi tiết của dữ liệu huấn luyện.

# 3. Thực hiện giải pháp

# Sau khi đề ra giải pháp, cần thực hiện giải pháp đó bằng cách huấn luyện lại mô hình với dữ liệu mới, kiến trúc mô hình mới hoặc các hyperparameter mới.

# 4. Đánh giá

# Cuối cùng, cần đánh giá xem giải pháp đã cải thiện độ chính xác của mô hình hay chưa. Điều này có thể được thực hiện bằng cách đánh giá mô hình trên tập dữ liệu kiểm tra. Nếu độ chính xác của mô hình đã cải thiện, thì giải pháp đã thành công. Ngược lại, nếu độ chính xác của mô hình không cải thiện hoặc thậm chí giảm xuống, thì cần xem xét lại giải pháp hoặc thực hiện các thay đổi khác.

# Một số ví dụ cụ thể về các giải pháp có thể được áp dụng để cải thiện độ chính xác của mô hình:

# * Trong trường hợp lỗi do dữ liệu:
#     - Có thể thu thập thêm dữ liệu từ các nguồn khác nhau.
#     - Có thể sử dụng các kỹ thuật xử lý dữ liệu để làm sạch dữ liệu, loại bỏ các mẫu dữ liệu nhiễu hoặc không phù hợp.
#     - Có thể sử dụng các kỹ thuật cân bằng dữ liệu để giải quyết vấn đề mất cân bằng dữ liệu.
# * Trong trường hợp lỗi do kiến trúc mô hình:
#     - Có thể thay đổi kiến trúc mô hình để phức tạp hơn, chẳng hạn như sử dụng các mạng nơ-ron sâu hơn.
#     - Có thể sử dụng các kỹ thuật regularize để ngăn ngừa overfitting.
#     - Có thể sử dụng các kỹ thuật ensemble để kết hợp dự đoán của nhiều mô hình khác nhau.
# * Trong trường hợp lỗi do các hyperparameter:
#     - Có thể thay đổi các hyperparameter liên quan đến số lượng vòng lặp huấn luyện, tốc độ học tập, tỷ lệ dropout, v.v.

# Có thể thấy, việc cải thiện độ chính xác của mô hình sau khi huấn luyện là một quá trình thử nghiệm và sai sót. Cần thử nghiệm nhiều phương pháp khác nhau để tìm ra phương pháp phù hợp nhất với bài toán của mình.
```
